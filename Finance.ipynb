```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.chdir("C:/Users/User/Documents/Empirial Stock Market Analysis/Pharmacy Developed")
```

```python
start_date = '2005-12-31'
end_date = '2016-01-01'
```

```python
data = {}
for f in os.listdir():
    file = pd.read_csv(f)
    file['Date'] = pd.to_datetime(file['Date'])
    file = file[ (file['Date'] > start_date) & (file['Date'] < end_date)]
    file.index = file['Date']
    del file['Date']
    data[f] = file 
data = dict(sorted(data.items(), key=lambda kv: len(kv[1]), reverse = True))
```

```python
plt.figure(figsize=(20,10))
avg = pd.DataFrame()
for k, v in data.items():
    v['Close'].plot(label = k.split('.')[0], style = ':')
    avg[k] = v['Close']
    if len(v.index) >= len(avg.index):
        avg.index = v.index
avg.dropna(inplace = True)
avg['mean'] = avg.mean(axis=1)
avg['mean'].plot(label = 'Average', lw = 4, color = 'r')
plt.axvline(x='2007-07-01', color = 'blue', label = 'Financial Crisis Start', ls = '-.', lw = 2)
plt.axvline(x='2010-01-01', color = 'black', label = 'Financial Crisis End', ls = '-.', lw = 2)
plt.title('Major Pharma Companies')
plt.legend()
```

```python
ovr = {}
ovr['Pharma'] = avg
```
